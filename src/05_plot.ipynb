```python
import matplotlib.pyplot as plt
```

```python
# Load fidelity data
fid_data = []
for i in range(100):
    with open("../scratch/fid-outputs/fidout_" + str(i) + ".txt", "r") as f:
        fid_data.append(f.readline().split(","))
for row in fid_data:
    row[0] = int(row[0])
    row[1] = float(row[1])
    row[2] = float(row[2])
    row[3] = float(row[3])
    row[4] = float(row[4])
```

```python
angles = [r[1] for r in fid_data]
data_po = [r[2] for r in fid_data]
data_pg = [r[3] for r in fid_data]
data_og = [r[4] for r in fid_data]
```

```python
plt.rcParams["text.usetex"] = True
```

```python
plt.plot(angles, data_po, label="Predicted-Optimized")
plt.plot(angles, data_pg, label="Predicted-Golden")
plt.plot(angles, data_og, label="Optimized-Golden")
plt.title("Gate Fidelity")
plt.xlabel("Gate Rotation Angle")
plt.ylabel("Fidelity")
plt.legend()
plt.show()
```
